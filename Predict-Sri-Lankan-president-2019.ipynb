## Bootstrapping

#### Removing votes marked as "No" and "Maybe"

```r
cleanedData <- subset(rawData, (isVoting != 'No' ))
cleanedData <- subset(cleanedData, (isVoting != 'Maybe' ))
```

```r
foo <- function(data, indices){
    dt <- data[indices,]    
    tbl <- table(dt$vote)
    t = as.data.frame(tbl)
    t[,2] <- t[,2]/sum(t[,2])*100
    c(mean(t[4,2]))
}
```

```r
library(boot)

set.seed(12345)
myBootstrap <- boot(cleanedData, foo, R=1000)
plot(myBootstrap, index=1)
```

###  Calculate 95% confident interval for Gotabaya Rajapaksha's vote percentage

```r
boot.ci(myBootstrap, index=1, type='norm')
```

#### We can conclude in 95% confident that Gotabaya Rajapaksha will get 66.95% to 89.75% percentage of votes from IT sector employees in this presidential election!!

## Bootstrapping data only for Colombo District

```r
cleanedDataColombo <- subset(cleanedData, (district == 'Colombo'))
```

```r
library(boot)

set.seed(12345)
myBootstrapColombo <- boot(cleanedDataColombo, foo, R=1000)
plot(myBootstrapColombo, index=1)
```

###  Calculate 95% confident interval for Gotabaya Rajapaksha's vote percentage

```r
boot.ci(myBootstrapColombo, index=1, type='norm')
```

#### We can conclude in 95% confident that Gotabaya Rajapaksha will get 47.84% to 89.75% percentage of votes from IT sector employees from Colombo District in this presidential election!!
